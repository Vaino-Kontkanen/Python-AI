Draw the lines y = 2x + 1, y = 2x + 2 and y = 2x + 3 in 
the same figure. Use different drawing colors and line 
types for your graphs to make them stand out in black and white. 
Set the image title and captions for the horizontal and vertical axes.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x = np.linspace(-5,5,1000)
plt.plot(x, 2*x+1, '-r', label='y=2x+1')
plt.plot(x,2*x+2,'--m', label='y=2x+2')
plt.plot(x,2*x+3, ':b', label='y=2x+3')
plt.legend(loc='upper left')
plt.title('Drawing Lines')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
